```python
import pandas as pd
import numpy as np
```

```python
# 载入数据
df = pd.read_excel('./test.xlsx',sheet_name='data')
df
```

```python
# 计算多条件众数
df_group = df.groupby(['id','year'])['num'].agg(lambda x:np.mean(x.mode())).reset_index()
df_group
```

```python
# 一维表转二维表
df_temp = pd.Series(list(df_group['num']),index=[df_group['id'],df_group['year']])  # 利用一维数据表 df_group 构造一个二级索引 Series 对象
df_result = df_temp.unstack().reset_index() # 对 df_temp 执行 stack 的逆操作 unstack
df_result
```

```python
# 保存数据
df_result.to_excel('./result.xlsx',sheet_name='result',index=False)
print(f'导出 Excel 文件成功！！')
```
