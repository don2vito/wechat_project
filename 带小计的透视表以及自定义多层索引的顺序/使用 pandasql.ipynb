```python
import pandas as pd
from pandasql import sqldf
```

```python
df = pd.read_excel('./test.xlsx',sheet_name='Sheet1')
df
```

```python
# pandasql 的底层是基于 SQLite 的

pyqldf = lambda q: sqldf(q, globals())

query1 = """
        SELECT
        df."地域",
        df."省份",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."数量"  ELSE 0 END ) AS "2020 数量",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."数量"  ELSE 0 END ) AS "2021 数量",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."数量"  ELSE 0 END ) AS "2022 数量",
        SUM (df."数量" ) AS "总计 数量",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."金额"  ELSE 0 END ) AS "2020 金额",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."金额"  ELSE 0 END ) AS "2021 金额",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."金额"  ELSE 0 END ) AS "2022 金额",
        SUM (df."金额" ) AS "总计 金额",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."市值"  ELSE 0 END ) AS "2020 市值",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."市值"  ELSE 0 END ) AS "2021 市值",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."市值"  ELSE 0 END ) AS "2022 市值", 
        SUM (df."市值" ) AS "总计 市值"
        FROM
        df 
        GROUP BY
        df."地域", df."省份"
        ORDER BY
        df."地域",
        df."省份"
        ;
        """

query2 = """
        SELECT
        df."地域",
        '小计' AS "省份",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."数量"  ELSE 0 END ) AS "2020 数量",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."数量"  ELSE 0 END ) AS "2021 数量",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."数量"  ELSE 0 END ) AS "2022 数量",
        SUM (df."数量" ) AS "总计 数量",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."金额"  ELSE 0 END ) AS "2020 金额",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."金额"  ELSE 0 END ) AS "2021 金额",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."金额"  ELSE 0 END ) AS "2022 金额",
        SUM (df."金额" ) AS "总计 金额",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."市值"  ELSE 0 END ) AS "2020 市值",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."市值"  ELSE 0 END ) AS "2021 市值",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."市值"  ELSE 0 END ) AS "2022 市值", 
        SUM (df."市值" ) AS "总计 市值"
        FROM
        df 
        GROUP BY
        df."地域"
        ORDER BY
        df."地域"
        ;
        """

query3 = """
        SELECT
        '总计' AS "地域",
        '' AS "省份",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."数量"  ELSE 0 END ) AS "2020 数量",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."数量"  ELSE 0 END ) AS "2021 数量",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."数量"  ELSE 0 END ) AS "2022 数量",
        SUM (df."数量" ) AS "总计 数量",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."金额"  ELSE 0 END ) AS "2020 金额",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."金额"  ELSE 0 END ) AS "2021 金额",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."金额"  ELSE 0 END ) AS "2022 金额",
        SUM (df."金额" ) AS "总计 金额",
        SUM ( CASE WHEN df."产品年"  = 2020 THEN df."市值"  ELSE 0 END ) AS "2020 市值",
        SUM ( CASE WHEN df."产品年"  = 2021 THEN df."市值"  ELSE 0 END ) AS "2021 市值",
        SUM ( CASE WHEN df."产品年"  = 2022 THEN df."市值"  ELSE 0 END ) AS "2022 市值",
        SUM (df."市值" ) AS "总计 市值"
        FROM
        df
        ;
        """

# 曲线救国，代替 UNION 的功能
result1 = pyqldf(query1)
result2 = pyqldf(query2)
result3 = pyqldf(query3)
result = pd.concat([result1,result2,result3])
result
```
