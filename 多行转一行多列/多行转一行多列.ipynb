```python
import pandas as pd
```

```python
# 读取文件
df_raw = pd.read_table('./实例文本.txt',sep=' ',header=None, encoding='gbk')
df_raw
```

```python
# 转置并删除空值
df_raw = df_raw.T
df_raw = df_raw.dropna(axis=0).reset_index()
df_raw.drop(['index'], axis=1, inplace=True) 
df_raw
```

```python
# 添加辅助列
num = 2
epoch = int(len(df_raw) / num)
temp_list = sorted(list(i for i in range(epoch)) * 2) # 升序排列

df_raw['辅助列'] = temp_list
df_raw.rename(columns={0:'内容'},inplace=True)
df_raw['内容'] = df_raw['内容'].astype(str)
df_raw
```

```python
# 分组合并
def func(df):
    return '/'.join(df.values)

df = df_raw.groupby(by='辅助列')['内容'].apply(func).reset_index()
df
```

```python
# 分列
df['编号'] = df['内容'].str.split('/',expand=True)[0]
df['规格'] = df['内容'].str.split('/',expand=True)[1]
df.drop(columns=['辅助列','内容'], axis=1, inplace=True)
df
```

```python
# 输出结果文件
df.to_excel('./result.xlsx',sheet_name='result',index=False)
print('生成结果文件成功！！')
```
