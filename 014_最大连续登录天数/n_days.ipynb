```python
import pandas as pd
import numpy as np
```

```python
# 数据去重
df = pd.read_excel('./实例文件.xlsx',sheet_name='数据')
df = df.drop_duplicates()
df
```

```python
# 组内排序
df['rank'] = df['login_time'].groupby(df['uid']).rank(method='dense',ascending=True).astype(int)
df = df.sort_values(by=['uid','login_time'])
df
```

```python
# 计算中间值
df['days'] = df['rank'].map(lambda x:pd.Timedelta(days=x))
df['minus'] = df['login_time'] - df['days']
df
```

```python
# 分组
df_group1 = df.groupby(['uid','minus'])['rank'].max().reset_index()
df_group1
```

```python
# 最终结果
df_group2 = df_group1.groupby('uid')['rank'].max().reset_index()
df_group2
```
