```python
import pandas as pd
```

```python
# 载入原始数据
df_raw = pd.read_excel('./实例文件.xlsx',sheet_name='Sheet1')
df_raw
```

```python
# 先转置一下
df_raw = df_raw.T
df_raw
```

```python
# 跳过表头 前 7 行
df_raw = df_raw.iloc[7:].reset_index()
df_raw
```

```python
# 每 10 行作循环，提取前 4 行，添加列，并进行拼接
df = pd.DataFrame()
totol_rows = len(df_raw)
nums = totol_rows // 10
rows_format = 10 * nums
new_list = [[i,i + 10] for i in range(0,rows_format,10)]
for i_j,num in zip(new_list,range(nums)):
    i,j = i_j
    temp_df = df_raw[i:j].iloc[:4]
    temp_df['分类'] = ['需求规模','可信度','价格合理性','新鲜感']
    temp_df['商品号'] = list(str(num + 1) * 4)
    df = df.append(temp_df)
df  
```

```python
# 逆透视
df = df.drop(columns='index')
df_melt = pd.melt(df,id_vars=['商品号','分类'])
df_melt['value'] = df_melt['value'].astype(float)
df_melt
```

```python
# 再透视，保留 2 位小数
df_pivot = pd.pivot_table(df_melt,index=['商品号'],columns=['分类'],values=['value'],aggfunc=['mean'],fill_value=0).applymap(lambda x:round(x,2))
df_pivot
```

```python
# 拍平多层索引，输出结果
df_pivot.columns = ['_'.join(col) for col in df_pivot.columns.values]
df_pivot.to_excel('./result.xlsx',sheet_name='result')
print('生成结果文件成功！！')
```
