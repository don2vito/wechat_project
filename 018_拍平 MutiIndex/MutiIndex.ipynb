```python
import pandas as pd
```

```python
# 读取数据
df = pd.read_excel('./sample.xlsx',sheet_name='data')
df
```

```python
# 透视表
import numpy as np

df_pivot = pd.pivot_table(df,index=['group','name'],values=['sale','price'],aggfunc=[np.sum,np.mean])
df_pivot
```

```python
# 生成 Excel 文件
df_pivot.to_excel('./result.xlsx',sheet_name='reslut')
print('生成 Excel 文件成功！！')
```

```python
# 生成 Excel 文件，使用 index=False 参数时报错
df_pivot.to_excel('./result.xlsx',sheet_name='reslut',index=False)
```

```python
df_pivot_join = df_pivot
df_pivot_merge = df_pivot
```

```python
# 拍平 MultiIndex
# 方法一：使用 join() 函数
df_pivot_join.columns = ['_'.join(col) for col in df_pivot_join.columns.values]
df_pivot_join
```

```python
# 拍平 MultiIndex
# 方法二：使用 '+' 连接
df_pivot_merge.columns = [col1 + '_' + col2 for (col1,col2) in df_pivot_merge.columns.tolist()]
df_pivot_merge
```

```python
# 再次输出 Excel 文件
df_pivot_merge.to_excel('./after_result.xlsx',sheet_name='after_reslut')
print('再次生成 Excel 文件成功！！')
```
