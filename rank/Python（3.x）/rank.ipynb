```python
import pandas as pd
```

```python
df = pd.read_excel('../toy_dataset.xlsx',sheet_name='raw_data')
df
```

```python
df['国际通用排名'] = df['数值'].groupby(df['文本列1']).rank(method='min',ascending=False).astype(int)
df['中国式排名'] = df['数值'].groupby(df['文本列1']).rank(method='dense',ascending=False).astype(int)
df
```

```python
df.to_excel('./result_python.xlsx',index=False)
```
