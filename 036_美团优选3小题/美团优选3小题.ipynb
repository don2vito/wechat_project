```python
import pandas as pd
```

```python
# 读取数据集
df1 = pd.read_excel('./示例数据.xlsx',sheet_name='data1')
df1
```

```python
df2 = pd.read_excel('./示例数据.xlsx',sheet_name='data2')
df2
```

```python
# 筛选出总交易额大于 500 的省份
df_merge = pd.merge(df1,df2,on='city_num',how='left')
df_group = df_merge.groupby(['province_name'])['gmv'].sum().reset_index()
df_group = df_group[df_group['gmv'] > 500]
df_province = df_group[['province_name']]
df_province
```

```python
# 得出省份总交易额在 [0,500)、[500,1000)、[1000,+无穷) 三个区间的省份数量
df_group = df_merge.groupby(['province_name'])['gmv'].sum().reset_index()
df_group['cut'] = pd.cut(df_group['gmv'],[0,500,1000,9999999], labels=['0 - 500','500 - 1000','1000 - 正无穷'])
df = df_group.groupby(['cut'])['province_name'].count().reset_index()
df
```

```python
# 按从小到大顺序得出每个城市的累计交易额
df_group = df_merge.groupby(['province_name'])['gmv'].sum().reset_index()
df_sort = df_group.sort_values(by=['gmv'],ascending=True)
df_sort['gmv_cumsum'] = df_sort['gmv'].cumsum()
df_sort
```
