```python
import pandas as pd
import datetime
import copy
```

```python
# 读取数据集
df = pd.read_excel('./data.xlsx',sheet_name='Sheet3')
df['日期'] = pd.to_datetime(df['日期'])
df = df.set_index('日期')
df
```

```python
df1 = copy.deepcopy(df)
df2 = copy.deepcopy(df)
df3 = copy.deepcopy(df)
```

```python
# 使用 shift() 移动数据
# 向下移动 2 行数据
df1 = df1.shift(periods=2)
df1
```

```python
# 向后移动 2 天索引（必须是时间索引）
df2 = df2.shift(freq=datetime.timedelta(2))
df2
```

```python
# 向后移动到月底索引（必须是时间索引）
df3 = df3.shift(freq='M')
df3
```

```python
# 使用 rolling() 移动计算数据
# 计算 3 日移动平均值
df = df.rolling(3).mean()
df
```

### rolling() 的各种计算方法
- rolling.count()  计算非空观察数
- rolling.sum（） 值的总和
- rolling.mean（） 平均值
- rolling.median（） 计算中值
- rolling.min（） 得到最小值
- rolling.max（） 得到最大值
- rolling.std（） 求贝塞尔修正样本标准差
- rolling.var（） 求无偏方差
- rolling.skew（） 样本偏度
- rolling.kurt（） 样本峰度
- rolling.quantile（） 样本分位数（参数值为百分比）
- rolling.apply（） 可以定义函数来灵活运用。
- rolling.cov（） 无偏协方差在数据处理中用于衡量两个变量的总体误差。
- rolling.corr（） 相关系数， 它是研究变量之间线性相关程度的量
